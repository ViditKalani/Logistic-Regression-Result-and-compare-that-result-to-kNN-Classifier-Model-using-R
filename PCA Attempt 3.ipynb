```python
from sklearn.datasets import fetch_mldata
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("German_Credit.csv", index_col = 0)
df.head()
```

```python
df['Class'].replace('Good', 1, inplace = True)
df['Class'].replace('Bad', 0, inplace = True)
df2 = df.drop(['Class'], axis = 1)
df2.head()
df3 = df[['Class']]
df3.head()
#df['Class']
```

```python
pca = PCA(n_components=10)
pca.fit(df2)
```

```python
print(pca.components_)
```

```python
principle_components = pca.fit_transform(df)
print(principle_components)
```

```python
df4 = pd.DataFrame(principle_components)
#df4.head()
df4.index = np.arange(1, len(df4) + 1)
df4.head()
df5 = df4.join(df3)
df5.head()
#pd.concat()
```

```python
np.savetxt("temp_credit.csv", principle_components, delimiter=",")
```

```python
df5.to_csv('pca_results.csv')
```
